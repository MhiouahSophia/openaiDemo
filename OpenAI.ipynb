
WORD2VEC : Example


```python
from gensim.models import Word2Vec
import nltk
nltk.download('punkt')
# Sample sentences

sentences = ['man', 'woman', 'rich', 'poor', 'queen', 'king', 'fisherman', 'teacher', 'actress', 'actor', "morgan Stanley", "bank", "investment"]


# Tokenize sentences
tokenized_sentences = [nltk.word_tokenize(sentence.lower()) for sentence in sentences]

# Train Word2Vec model
model = Word2Vec(tokenized_sentences, vector_size=100, window=5, min_count=1, sg=0)

```

```python
# Initialize a zero matrix for embedding sentences
embedded_sentences = np.zeros((len(tokenized_sentences), model.vector_size))

# Replace tokens with word embeddings and fill the matrix
for i, sentence_tokens in enumerate(tokenized_sentences):
    valid_tokens = [token for token in sentence_tokens if token in model.wv]
    if valid_tokens:
        word_embeddings = model.wv[valid_tokens]
        sentence_embedding = np.mean(word_embeddings, axis=0)
        embedded_sentences[i] = sentence_embedding


# Create a t-SNE model and transform the data
tsne = TSNE(n_components=2, perplexity=1, random_state=42, init='random', learning_rate=200)
vis_dims = tsne.fit_transform(embedded_sentences)

x = [x for x,y in vis_dims]
y = [y for x,y in vis_dims]

plt.scatter(x, y,  alpha=0.3)

for i, txt in enumerate(sentences):
    plt.annotate(txt, (x[i], y[i]))

plt.title(" Visualizing in language using t-SNE")
```

```python
import pandas as pd
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import openai

openai.api_key = "secret key"
```

```python
text = ['man', 'woman', 'rich', 'poor', 'queen', 'king', 'fisherman', 'teacher', 'actress', 'actor']

embed = []
for i in text:
  embed.append(openai.Embedding.create(input = [i], model="text-embedding-ada-002")['data'][0]['embedding'])

matrix = np.array(embed)

# Create a t-SNE model and transform the data
tsne = TSNE(n_components=2, perplexity=3, random_state=42, init='random', learning_rate=200)
vis_dims = tsne.fit_transform(matrix)

x = [x for x,y in vis_dims]
y = [y for x,y in vis_dims]

plt.scatter(x, y,  alpha=0.3)

for i, txt in enumerate(text):
    plt.annotate(txt, (x[i], y[i]))

plt.title(" Visualizing in language using t-SNE")
```
